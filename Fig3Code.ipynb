```python
from scipy.stats import expon
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.stats import binom
import pandas as pd
from scipy.stats import norm

```

```python

data_expon = expon.rvs(scale=10,loc=0,size=1000)
ax = sns.distplot(data_expon,
                  kde=True,
                  bins=100,
                  color='black',
                  hist_kws={"linewidth": 15,'alpha':1, 'color':'lightgrey'})
ax.set(xlabel='Branch lengths drawn from exponential distribution', ylabel='Frequency')
```

```python
ax.figure.savefig("Fig3.png", dpi=500)
#fig.savefig("Fig3.png", dpi="500", facecolor='w', edgecolor='w')
```

```python
data_binom = binom.rvs(n=1,p=0.2,size=42)
```

```python
ax = sns.distplot(data_binom,
                  kde=False,
                  color='lightgrey',
                  bins=2,
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set_xlabel('Character Frequency', fontsize=18)
ax.set_ylabel('Frequency', fontsize=18)
```

```python
fig = ax.get_figure()
fig.savefig('Fig5.png', dpi=500)
```

```python
data_norm = norm.rvs(size=10000,loc=4,scale=.5)
```
